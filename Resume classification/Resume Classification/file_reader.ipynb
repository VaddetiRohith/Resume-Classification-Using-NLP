```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import docx2txt
import glob
import re
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
```

***Pepplesoft***

```python
data = []
file_names = []

# Use glob to get paths of all .doc and .docx files in folder
files = glob.glob(
    r"D:\ExcelR Project\Resume classification\Resumes\Peoplesoft resumes\*.docx"
) + glob.glob(
    r"D:\ExcelR Project\Resume classification\Resumes\Peoplesoft resumes\*.doc"
)

# Loop through files
for file in files:
    # Get file name and append to file_names list
    file_name = file.split("\\")[-1]
    file_names.append(file_name)

    # Read text data from file and append to data list
    if file_name.endswith('.docx'):
        text = docx2txt.process(file)
        data.append(text)
    else:
        with open(file, 'rb') as f:
            text = str(f.read())
            data.append(text)

# Create a dataframe
df = pd.DataFrame({'Resume': data})
```
